```python
# part 1 - fetching the required stock data for the 10 selected stocks
#AAPL
AAPL_df = yf.download(['AAPL'],     #Using yfinance we use the download function to fetch and save the indicated stock to a df
start='2012-11-15',  #start date for stocks
end= '2022-11-15', #end date for stocks
progress=False)

#MSFT
MSFT_df = yf.download(['MSFT'],
start='2012-11-15',
end= '2022-11-15',
progress=False)
MSFT_High = MSFT_df
MSFT_Low = MSFT_df
#AMZN
AMZN_df = yf.download(['AMZN'],
start='2012-11-15',
end= '2022-11-15',
progress=False)

#TSLA
TSLA_df = yf.download(['TSLA'],
start='2012-11-15',
end= '2022-11-15',
progress=False)
TSLA_High = TSLA_df
TSLA_Low = TSLA_df
#GOOGL
GOOGL_df = yf.download(['GOOGL'],
start='2012-11-15',
end= '2022-11-15',
progress=False)

#META
META_df = yf.download(['META'],
start='2012-11-15',
end= '2022-11-15',
progress=False)

#NVDA
NVDA_df = yf.download(['NVDA'],
start='2012-11-15',
end= '2022-11-15',
progress=False)

#PEP
PEP_df = yf.download(['PEP'],
start='2012-11-15',
end= '2022-11-15',
progress=False)

#COST
COST_df = yf.download(['COST'],
start='2012-11-15',
end= '2022-11-15',
progress=False)

#AVGO
AVGO_df = yf.download(['AVGO'],
start='2012-11-15',
end= '2022-11-15',
progress=False)
```

```python
#loading backtesting library which will be used for the majority of the simulations and analytics do to it's handy testing tools
from backtesting import Backtest, Strategy
from backtesting.lib import crossover

import talib as ta
```

```python

def MACD(close, fast, slow, signal):
  #This function will be used to assign macd and signal values to individual stocks based on the strategy given
    macd, macdsignal, macdhist = ta.MACD(close, fastperiod=fast, slowperiod=slow, signalperiod=signal)
    return macd, macdsignal

#object created based on backtesting documentation 
class MACDCross(Strategy):
 #Assigning macd values 
  fast = 12 
  slow = 26 
  signal = 9 
  #mapping out the macd and macd signal
  def init(self):
    self.macd, self.macdsignal = self.I(MACD, self.data.Close, self.fast, self.slow, self.signal)
  #Implementing our trategy
  def next(self): 
    if crossover(self.macd, self.macdsignal): #buying when macd crosses up on macd signal
        self.buy() 
    elif crossover(self.macdsignal, self.macd): #selling when macd crosses down on macd signal
        self.position.close() 
```

```python
#We now will use the Backtest formula to find out how our strategy will do in the duration of our dataframe
bt_TSLA = Backtest(
    TSLA_df,
    MACDCross, #using our strategy
    cash=10000, #Starting cash
    commission=0,  #set comission to 0 as we assume zero fee trading, we can also change to 0.05 for a more realistic simulation
    exclusive_orders=True
)

bt_AAPL = Backtest(
    AAPL_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)

bt_MSFT = Backtest(
    MSFT_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)

bt_AMZN = Backtest(
    AMZN_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)

bt_GOOGL = Backtest(
    GOOGL_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)

bt_META = Backtest(
    META_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)

bt_NVDA = Backtest(
    NVDA_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)

bt_PEP = Backtest(
    PEP_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)

bt_COST = Backtest(
    COST_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)

bt_AVGO = Backtest(
    AVGO_df,
    MACDCross, 
    cash=10000,
    commission=0,  
    exclusive_orders=True
)
```
